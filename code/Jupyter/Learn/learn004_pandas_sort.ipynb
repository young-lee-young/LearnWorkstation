```python
import pandas as pd
```

```python
dataframe = pd.read_csv('D:/spider/1.csv', encoding='gbk')
```

```python
dataframe
```

```python
# 行排序，默认是升序排序，ascending=False为降序排序

dataframe.sort_index(ascending=False)
```

```python
# 加上axis为列排序

dataframe.sort_index(axis=1)
```

```python
# 按两个列进行排序，用by=['列名', '列名']

dataframe.sort_values(['Sex', 'NickName'])
```

```python
series = pd.Series([2,4,5,3,1], index=['a', 'e', 'c', 'b', 'd'])
```

```python
series
```

```python
# Series按索引排序,排序是视图，不是在原数据上修改

series.sort_index()
```

```python
# Series按值排序

series.sort_values()
```

```python
# rank函数还是不太明白

series.rank()
```
