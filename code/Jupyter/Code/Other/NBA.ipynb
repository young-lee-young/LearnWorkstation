```python
#随机产生5个中心点
num_clusters = 5
random_initial_points = np.random.choice(nba.index,num_clusters)
centroids = nba.ix[random_initial_points]
print(centroids)
```

```python
plt.scatter(nba['score'],nba['assist'],c = 'yellow')
plt.scatter(centroids['score'],centroids['assist'],c = 'red')
plt.title("Centroids")
plt.xlabel('Points Per Game',fontsize = 13)
plt.ylabel('Assist Per Game',fontsize = 13)
```

```python
def centroids_to_dict(centroids):
    dictionary = dict()
    counter = 0
    print(type(centroids))
    for index,row in centroids.iterrows():
        coordinates = [row['score'],row['assist']]
        dictionary[counter] = coordinates
        counter +=1
    
    return dictionary

centroids_dict = centroids_to_dict(centroids)
print(centroids_dict)
```

```python
#计算两个点之间的距离
import math
def calculate_distance(centroid,players_values):
    root_distance=0

    for x in range(0,len(centroid)):
        difference = centroid[x]-players_values[x]
        squared_difference = difference**2
        root_distance += squared_difference
    euclid_distance = math.sqrt(root_distance)
    return euclid_distance

def assign_to_cluster(row):
    lowest_distance = -1
    closest_cluster = -1
    for cluster_id,centroid in centroids_dict.items():
        df_row = [row['score'],row['assist']]
        print(centroid,df_row)
        euclidean_distance = calculate_distance(centroid,df_row)
        print("欧几里得距离：",euclidean_distance)
        if lowest_distance == -1:
            lowest_distance = euclidean_distance
            closest_cluster = cluster_id
        elif euclidean_distance < lowest_distance:
            lowest_distance = euclidean_distance
            closest_cluster = cluster_id
    return closest_cluster

nba['cluster'] = nba.apply(lambda row:assign_to_cluster(row),axis = 1)
print(nba)
```

```python
def visualize_clusters(df,num_clusters):
    colors = ['b','g','r','c','m','y','k']
    
    for n in range (num_clusters):
        clustered_df = df[df['cluster'] == n]
        plt.scatter(clustered_df['score'],clustered_df['assist'],c = colors[n-1])
        plt.xlabel('Points Per Game',fontsize = 13)
        plt.ylabel('Assist Per Game',fontsize = 13)
    
visualize_clusters(nba,5)
```

```python
def recalculate_centroids(df):
    new_centroid_dict = dict()
    
    for cluster_id in range(0,num_clusters):
        values_in_cluster = df[df['cluster']==cluster_id]
        new_centroid = [np.average(values_in_cluster['score']),np.average(values_in_cluster['assist'])]
        new_centroid_dict[cluster_id] = new_centroid
    return new_centroid_dict
centroid_dict = recalculate_centroids(nba)
```

```python
#重新计算中心点
centroids_dict = recalculate_centroids(nba)
nba["cluster"] = nba.apply(lambda row:assign_to_cluster(row),axis=1)  
```

```python
visualize_clusters(nba,5)
```

```python
#以上代码是自己实现,以下是利用库函数实现
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(nba[['score','assist']])
nba['cluster'] = kmeans.labels_
visualize_clusters(nba,num_clusters)
```
